```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, segmentation, color, measure

# Read the grayscale image
image = io.imread('/Applications/ML projects/Image Colorisation/archive/indoorCVPR_09/Images/airport_inside/airport_inside_0001.jpg', as_gray=True)

regions = segmentation.slic(image, compactness=10, n_segments=100)

# Obtain region properties
props = measure.regionprops(regions)

# Assign random colors to each region
colors = np.random.randint(0, 256, (np.max(regions) + 1, 3), dtype=np.uint8)

# Create a colored image based on region labels
colored_image = colors[regions]

# Plotting the results
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(colored_image)
axes[1].set_title('Segmented Image')
axes[1].axis('off')
plt.tight_layout()
plt.show()
```
